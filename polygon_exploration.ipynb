```python
import folium
import json

# Define the file path to the GeoJSON file
geojson_file = "/home/user/MSc/Clean_Codebase/scripts/nile_polygon.geojson"

# Load GeoJSON data
with open(geojson_file, 'r') as file:
    nile_geojson = json.load(file)

# Create a Folium Map centered around the Nile Delta region
m = folium.Map(location=[31.5, 30.0], zoom_start=10)

# Add the GeoJSON overlay
folium.GeoJson(
    nile_geojson,
    name="Nile Delta",
    style_function=lambda feature: {
        'fillColor': 'blue',
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.5,
    }
).add_to(m)

# Add Layer Control
folium.LayerControl().add_to(m)

display(m)
```
